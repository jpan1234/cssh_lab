# Obtaining titles, urls, and dates from the main speeches webpage

In html, we can access multiple pieces of information within the main [speeches page](https://aflcio.org/speeches) of the AFLCIO website. We can find certain patterns within the html code to access the information using the beautiful soup package. For example, the html tag `<h2 class>` contains the speech title.

```python
# main speeches webpage 
# set the main url
URL_main = 'https://aflcio.org/speeches'
# request the url
r_main = requests.get(URL_main)
# open up with BeautifulSoup
soup = BeautifulSoup(r_main.content, 'html5lib')

# show the html code
print(soup)
```

## Finding Speech names, their dates, and their URLs 

We can find the speech names, dates, and URLs on the main page, as we loop through the pages and scrape their respective data, we can append them all to their respective lists

```python
# SPEECH NAMES 

# create speech names list
speech_names = []

# splits by <h2> to obtain the speech names
for data in soup.find_all("h2"):
    
    # speech title
    speech_title = data.get_text()
    
    # append the title to the names
    speech_names.append(speech_title)

# remove the breadcrumbs section
speech_names = speech_names[1:]
print(speech_names)


# create a table to find the urls
table = soup.find('div', attrs = {'class':'content-details'}) 
```

```python
# URLS


# make url list
url_list = []

# loop through the database for class b-inner
for row in table.find_all_next('a', attrs = {'class': 'b-inner'}):

    # url extension
    url = row['href']
    
    # append to url list
    url_list.append(url)

print(url_list)
```

```python
# make date list make url list
date_list = []

for data in soup.find_all('time'):
    date = data.get_text()
    date_list.append(date)

print(date_list)
```

```python
# check that the indexes are matching
for i in range(0,5):
    print(date_list[i])
    print(url_list[i])
    print(speech_names[i])
   
```
